```python
import numpy as np

# The following functions implement channel-by-channel and concatenated classifiers for MTS.
from sklearn.preprocessing import StandardScaler
def channelBYchannel_classifier(X_train, y_train, X_test, y_test, classifier, std=False, coeff=False):
    # function that applies a classifier channel-by-channel
    # returns as prediction the most frequent label for each test sample and the accuracy score
    # is possible to standardize each channel before classification
    # if coeff is True, it returns also the coefficient of the classifier (needs classifier to return coefficients, for example regressive classifiers)
    y_pred = np.zeros((X_test.shape[0], X_train.shape[1]), dtype=int)  # Initialize predictions array
    evaluate = np.zeros((X_train.shape[1]), dtype=float)  # Initialize scores array
    coef = np.zeros((X_train.shape[2], X_train.shape[1]), dtype=float)  # Initialize weights array
    X_train_nor = X_train.copy()
    X_test_nor = X_test.copy()
    for i in range(X_train.shape[1]):
        if std:
            # Normalize each channel
            X_train_nor[:,i,:] = StandardScaler().fit_transform(X_train[:,i,:])
            X_test_nor[:,i,:] = StandardScaler().fit_transform(X_test[:,i,:])
        # Predict and score for each channel
        if coeff:
            y_pred[:,i], evaluate[i], coef[:,i] = classifier(X_train_nor[:,i,:], y_train, X_test_nor[:, i, :], y_test)
        else:
            y_pred[:,i], evaluate[i], _ = classifier(X_train_nor[:,i,:], y_train, X_test_nor[:, i, :], y_test)
    y = np.zeros(X_test.shape[0], dtype=str)  # Initialize final predictions array
    for i in range(X_test.shape[0]):
        # Use the mode of predictions across channels for each test sample
        y[i] = np.bincount(y_pred[i,:]).argmax()
    score = np.mean(y == y_test)  # Calculate accuracy
    return y, score, evaluate, coef

def concatenated_classifier(X_train, y_train, X_test, y_test, classifier, std=False, coeff=False):
    # function tha applies a classifier concatenating the channels
    # is possible to standardize each channel before classification
    # if coeff is True, it returns also the coefficients of the classifier (needs classifier to return coefficients, for example regressive classifiers)
    coef = np.zeros((X_test.shape[2]), dtype=float)
    X_train_nor = X_train.copy()
    X_test_nor = X_test.copy()
    if std:
        for i in range(X_train.shape[1]):
            # Normalize each channel
            X_train_nor[:,i,:] = StandardScaler().fit_transform(X_train[:,i,:])
            X_test_nor[:,i,:] = StandardScaler().fit_transform(X_test[:,i,:])
    X_train_concat = np.zeros((X_train.shape[0], X_train.shape[1] * X_train.shape[2]), dtype=float)
    X_test_concat = np.zeros((X_test.shape[0], X_test.shape[1] * X_test.shape[2]), dtype=float)
    for i in range(X_train.shape[0]):
        X_train_concat[i,:] = np.concatenate(X_train_nor[i,:,:], axis=None)
        X_test_concat[i,:] = np.concatenate(X_test_nor[i,:,:], axis=None)
    if coeff:
        y, score, coef = classifier(X_train_concat, y_train, X_test_concat, y_test)
    else:
        y, score, _ = classifier(X_train_concat, y_train, X_test_concat, y_test)
    return y, score, coef

# The following functions implement various time series classification algorithms using classical methods.

from sktime.classification.distance_based import KNeighborsTimeSeriesClassifier
def knn_euclidean_classifier(X_train, y_train, X_test, y_test):
    # k-nearest neighbors classifier (k=1, distance = euclidean)
    classifier = KNeighborsTimeSeriesClassifier(distance="euclidean")
    classifier.fit(X_train, y_train)
    y = classifier.predict(X_test)
    score = np.mean(y == y_test)  # Calculate accuracy
    return y, score, _

from sklearn.linear_model import LogisticRegression
def logistic_classifier(X_train, y_train, X_test, y_test):
    # Logistic regression classifier
    classifier = LogisticRegression(penalty=None)
    classifier.fit(X_train, y_train)
    y = classifier.predict(X_test)
    score = np.mean(y == y_test)  # Calculate accuracy
    return y, score, classifier.coef_

from sklearn.linear_model import RidgeClassifierCV
def ridge_classifier(X_train, y_train, X_test, y_test):
    # Ridge regression classifier with cross-validation
    classifier = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10), store_cv_results=True, scoring='accuracy')
    classifier.fit(X_train, y_train)
    print("Ridge classifier best alpha:", classifier.alpha_)
    y = classifier.predict(X_test)
    score = np.mean(y == y_test)  # Calculate accuracy
    return y, score, classifier.coef_

# The following function implement a specific TSC algorithms from sktime library.

from sktime.classification.kernel_based import RocketClassifier
def rocket_classifier(X_train, y_train, X_test, y_test):
    # ROCKET classifier (random convolutional kernel-based classifier)
    classifier = RocketClassifier(num_kernels=5000, n_jobs=6)  # Use multiple jobs for faster computation
    classifier.fit(X_train, y_train)
    y = classifier.predict(X_test)
    score = np.mean(y == y_test)  # Calculate accuracy
    return y, score, _
```
